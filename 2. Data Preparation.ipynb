###**a) Lakukan Integrasi data pada kedua dataset yaitu dataset diabetes_RS1 dan diabetes_RS2 dengan memperhatikan kolom Age ! (Nb. Penulisan nama kolom harus sama persis, maka ubah salah satu nama kolom terlebih dahulu menjadi Age atau AGE)**

```python
import pandas as pd                         #lakukan import library untuk perhitungan numerik
dataset1 = "/content/diabetes_RS1.csv"      #Read dataset csv yang digunakan
dataset2 = "/content/diabetes_RS2.csv"
df1 = pd.read_csv(dataset1,header = 0)      #Proses pembacaan csv
df2 = pd.read_csv(dataset2,header = 0)
```

Pertama kita cek dahulu atribut kolom yang ada pada dataset diabetes 1 dan diabetes 2 untuk melihat kolom atribut yang sama


```python
df1.head()
```

```python
df2.head()
```

Dari kedua dataset tersebut, terdapat satu kolom yang memiliki keterangan dan data yang sama yaitu pada atribut Age(umur). \
Sebelum menggabungkan kolom, dilakukan perubahan nama kolom pada salah satu dataset. Dilakukan pada dataset 1



```python
df1.rename(columns={"Age":"AGE"}, inplace=True)
df1.head()
```

Kemudian lakukan integrasi \\
Sehingga tercipta dataset baru yang dimasukkan ke dalam variabel df

```python
df = pd.merge(df1, df2, on = 'AGE')
df.head(10)
```

save CSV terbarunya!

```python
df.to_csv("data_integration_damin2.csv")
```

#### Transformasi Data Encoding

```python
import pandas as pd
import numpy as np
df = pd.read_csv('/content/Student_bucketing.csv', header = 0) # Membaca file csv pada dataset mark, baris pertama dimulai dari 0
df.head() #Menampilkan 5 baris teratas
```

```python
df_categorical['Employed'].unique() ##Menampilkan nilai unik pada dfkategori kolom Employed
```

```python
df_categorical.head(10) #Menampilkan 10 baris pertama data categori
```

#### Transformasi Data Label Encoding

```python
import numpy as np
dataset = "/content/Banking_Marketing.csv" #membuat variabel dtaset yang berisi dataset
df = pd.read_csv(dataset, header=0) #membaca file csv dan baris dimulai dari 0
```

```python
df=df.dropna() #menghapus semua missing value
```

```python
data_kolom = df.select_dtypes(exclude=[np.number]).columns # Membuat df baru berisi kolom non-numerik dengan mengecualikan kolom numerik menggunakan library numpy
data_kolom
```

```python
df[data_kolom].head() # Menampilkan 5 baris pertama dataframe yg berisi kolom non numerik
```

```python
from sklearn.preprocessing import LabelEncoder # Mengimport LabelEncoder dari modul preprocessing dalam library Scikit-Learn

label_encoder = LabelEncoder() # Membuat objek instance dari LabelEncoder

for i in data_kolom: # Perulangan i pada data kolom / melakukan iterasi
  df[i] = label_encoder.fit_transform(df[i]) # Di lakukan transformasi menggunakan objek instance sebelumnya

print("Label Encode Data:") # Output
df.head() # Menampilkan 5 baris pertama data frame
```

```python
df.to_csv("data_transformation_labeling_encoding_damin2.csv")
```

**Analisis :**\
Pada trasformasi data label encoding digunakan dataset Banking_Marketing.csv. Pada transformasi data label encoding harus dilakukan missing value terlebih dahulu. Kemudian pemmisahan data non numerik dan terdapat 10 atribut non numerik. Kemudian data tersebut dijadikan menjadi satu df. Untuk melakukan label encoding maka perlu LabelEncoder dari modul preprocessing dalam Library Scikit-Learn. Lalu, dilakukan transformasi data non numerik dengan perulangan atau iterasi menjaddi data numerik.

###**d) Dari semua dataset yang ada di praktikum kali ini pilihlah minimal satu dataset kemudian lakukan Data Discretization!**

```python
import pandas as pd
dataset = "/content/mark.csv" #mendefinisikan dataset
df = pd.read_csv(dataset, header = 0) #membaca file csv dengan dimulai dari 0
```

```python
df.head() #menampilkan 5 baris dataset teratas
```

```python
# Lakukan pengelompokan menggunakan fungsi pd.cut() pada kolom marks
df['bucket'] = pd.cut(df['Mark'],3, labels=['Poor', 'Average', 'Excellent']) #Melakukan pengelompokan dengan 3 variabel yaitu poor, average dan excellent pada mark
df.head(10) # Menampikan 10 baris pertama
```

```python
df.to_csv("data_Discretization_damin2.csv") #menyimpan file csv
```

**Analisis :**\
Pada data discretization digunakan dataset mark.csv. Pada data discretization dilakukan pengelompokan fungsi pada mark dengan memberi 3 label menggunakan fungsi pd.cut(). Kemudian dapat dilihat hasil dari proses ini.
